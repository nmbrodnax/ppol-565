```python
# Visualization with Python
# NaLette Brodnax
# PPOL 565, Spring 2019

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import data
counties = pd.read_csv('../data/county_data.csv')
counties.head()
```

```python
# Histogram
plt.hist(counties['adult_pop'], bins=50)
plt.show()
```

```python
# Histogram of transformed scale
counties['log_pop'] = np.log(counties['adult_pop'])
plt.hist(counties.log_pop, bins=50)
plt.show()
```

```python
# Always add title and labels
plt.hist(counties['log_pop'], bins=50)
plt.title('Histogram of Adult Population')
plt.xlabel('Log Adult Population')
plt.ylabel('Count')
```

```python
# Scatterplot 
plt.scatter(counties.log_pop, counties.median_income)
plt.title('Median Income by County Size')
plt.xlabel('Log Adult Population')
plt.ylabel('Median Income')
```

```python
# Scatterplot with transformed scale
counties['median_income_k'] = counties['median_income']/1000
plt.scatter(counties.log_pop, counties.median_income_k)
plt.title('Median Income by County Size')
plt.xlabel('Log Adult Population')
plt.ylabel('Median Income (K)')

# Save the file
plt.savefig('county_inc_size.pdf')
```
